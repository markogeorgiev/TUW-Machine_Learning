# Student's Academic Success Data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

students_success_data_raw = pd.read_csv('students_academic_success_data.csv', sep=';')
students_success_data_raw.head()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Target', data=students_success_data_raw, palette='Set2')
plt.title('Distribution of "Target"')
plt.xlabel('Target')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=students_success_data_raw, x='Age at enrollment', bins=20, kde=True, color='skyblue')
plt.title('Distribution of Age at enrollment')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

X_students = students_success_data_raw.drop('Target', axis=1)
y_students = students_success_data_raw['Target']

X_train, X_test, y_train, y_test = train_test_split(X_students, y_students, test_size=0.2, random_state=42)

model = RandomForestClassifier(n_estimators=1000, random_state=42)
model.fit(X_train, y_train)

model.score(X_test, y_test)
```
